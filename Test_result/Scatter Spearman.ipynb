```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
```

```python
file_name = "spatial_recognition_responses.csv"
sr_r = pd.read_csv("spatial_recognition_responses.csv")
file_name= "ansrespclean.csv"
ans_r = pd.read_csv("ansrespclean.csv")
print (ans_r)


```

```python
# ANS score vs spatial score

result = pd.read_csv("ANS vs Spatial.csv")

x=result['y']
y=result['x']


plt.scatter(x,y)

slope, intercept = np.polyfit(x, y, 1)
x_line = np.array([min(x), max(x)])
y_line = slope * x_line + intercept

plt.plot(x_line,y_line,color="red")

plt.xlabel
plt.ylabel
plt.title("ANS score vs spatial recognition score")

r_value = np.corrcoef(x, y)[0, 1]

plt.xlabel('Spatial recognition score')
plt.ylabel('ANS score')
plt.title('Scatter Plot of ANS vs Spatial (r = {:.2f})'.format(r_value))

plt.show()


corr, pval = spearmanr(x, y)
print("Spearman's correlation coefficient:", corr)
print("p-value",pval)
```
