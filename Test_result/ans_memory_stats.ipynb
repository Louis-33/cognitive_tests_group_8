```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#read data from csv file
filename = "memory_test_responses.csv"
memory_df = pd.read_csv(filename)

#drop items that are not needed for the calculation
memory_df = memory_df.drop( [ 'Timestamp', 'name', 'age',  'Level 1 score', 'Level 2 score', 'Level 3 score', 'time_taken'], axis='columns' )
memory_df

#get males and females score from table respectively
memory_male_scores = memory_df[memory_df['gender']=='Male']['total_score']
memory_female_scores = memory_df[memory_df['gender']=='Female']['total_score']

#find mean
memory_m_mean = memory_male_scores.mean()
memory_f_mean = memory_female_scores.mean()

#find difference in mean
memory_diff_in_mean = memory_m_mean - memory_f_mean
print(memory_diff_in_mean)
#difference in mean = -1.96. Mean female total_score is 1.69 greater than mean male total_score

#calculate significant difference using t-test
results_memory = stats.ttest_ind(memory_male_scores, memory_female_scores)
results_memory.pvalue
#p-value = 0.13
```

```python
###intepreting the t-test p-value -> memory test <- ### 
#p-value = 0.13

# If p-value â‰¤ alpha:
# Reject the null hypothesis.
# Conclude that there is evidence to suggest a significant effect or difference.

# If p-value > alpha:
# Fail to reject the null hypothesis.
# Conclude that there is not enough evidence to suggest a significant effect or difference.

#alpha chosen = 0.05 (0.5%) 
#since p-value > 0.05
#null hypothesis can be accepted
#meaning that there is not enough evidence to suggest a significant effect or difference in the memory test of male and female.
```

```python
#create a box plot using Matplotlib
plt.figure(figsize=(8,6))
plt.boxplot([memory_male_scores, memory_female_scores], labels = ['Male','Female'])

#set labels and title
plt.ylabel('Score')
plt.title('Memory Test Box Plot')

#show the plot
plt.show()
```

```python
memory_std_scores = memory_df.groupby('gender')['total_score'].std()
memory_mean_scores = memory_df.groupby('gender')['total_score'].mean()
#calculate mean and standard deviation for males and females
```

```python
#create a dot plot with error bars
plt.figure(figsize=(8,6))
plt.scatter(memory_df['gender'], memory_df['total_score'], color = 'pink', label='Individual Scores')
plt.errorbar(memory_mean_scores.index, memory_mean_scores, yerr=memory_std_scores, fmt='o', color='grey', label='Mean Â± Std dev')
plt.ylabel('Score')
plt.title('Memory Test Dot Plot')
plt.legend()
plt.show()
```

```python
#read data from csv file
filename = "ans_responses.csv"
ans_df = pd.read_csv(filename)

#drop user id from the table
ans_df = ans_df.drop( [ 'user_id'], axis='columns')

#get male and female scores individually
```

```python
#get males and females score from table respectively
ans_male_scores = ans_df[ans_df['gender']=='M']['score']
ans_female_scores = ans_df[ans_df['gender']=='F']['score']

#find mean
ans_m_mean = ans_male_scores.mean()
ans_f_mean = ans_female_scores.mean()

#find difference in mean
ans_diff_in_mean = ans_m_mean - ans_f_mean
print(ans_diff_in_mean)
#difference in mean = 0.136. Mean male total_score is 0.136 greater than mean male total_score

#calculate significant difference using t-test
t_statistic, p_value = stats.ttest_ind(ans_male_scores, ans_female_scores)
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")
#p-value=0.922
```

```python
###intepreting the t-test p-value -> ans test <- ### 
#p-value = 0.922

# If p-value â‰¤ alpha:
# Reject the null hypothesis.
# Conclude that there is evidence to suggest a significant effect or difference.

# If p-value > alpha:
# Fail to reject the null hypothesis.
# Conclude that there is not enough evidence to suggest a significant effect or difference.

#alpha chosen = 0.05 (0.5%) 
#since p-value > 0.05
#null hypothesis can be accepted
#meaning that there is not enough evidence to suggest a significant effect or difference in the ans test of male and female.
```

```python
#create a box plot using Matplotlib
plt.figure(figsize=(8,6))
plt.boxplot([ans_male_scores, ans_female_scores], labels = ['Male','Female'])

#set labels and title
plt.ylabel('Score')
plt.title('ANS Test Box Plot')

#show the plot
plt.show()
```

```python
ans_std_scores = ans_df.groupby('gender')['score'].std()
ans_mean_scores = ans_df.groupby('gender')['score'].mean()
#calculate mean and standard deviation for males and females
```

```python
#create a dot plot with error bars
plt.figure(figsize=(8,6))
plt.scatter(ans_df['gender'], ans_df['score'], color = 'pink', label='Individual Scores')

#displaying error bars on the graph
plt.errorbar(ans_mean_scores.index, ans_mean_scores, yerr=ans_std_scores, fmt='o', color='grey', label='Mean Â± Std dev')

plt.ylabel('Score')
plt.title('ANS Test Dot Plot')
plt.legend()

#show dot plot
plt.show()
```
