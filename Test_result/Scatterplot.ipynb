```python
import pandas as pd
import matplotlib.pyplot as plt
# Create a figure with 4 subplots arranged in a 2x2 grid
fig, axs = plt.subplots(2,2)

ans = pd.read_csv('ansrespclean.csv')
spat = pd.read_csv('spatial_recognition_responses.csv')
math = pd.read_csv('math_ability_responses.csv')
memory = pd.read_csv('memory_test_responses.csv')

# Create a scatter plot for each feature in the data
axs[0, 0].scatter(ans['score'][0:15], spat['score'][0:15])
axs[0, 1].scatter(ans['score'][0:15], math['score'][0:15])
axs[1, 0].scatter(ans['score'][0:15], memory['total_score'][0:15])

# Set the axis labels and title for each subplot
axs[0, 0].set_xlabel('ans')
axs[0, 0].set_ylabel('spatial')
axs[0, 0].set_title('Scatter Plot of x1 vs y1')

axs[0, 1].set_xlabel('ans')
axs[0, 1].set_ylabel('math')
axs[0, 1].set_title('Scatter Plot of x2 vs y2')

axs[1, 0].set_xlabel('ans')
axs[1, 0].set_ylabel('memory')
axs[1, 0].set_title('Scatter Plot of x3 vs y3')

axs[1, 1].set_xlabel('x4')
axs[1, 1].set_ylabel('y4')
axs[1, 1].set_title('Scatter Plot of x4 vs y4')

# using padding
fig.tight_layout(pad=1.0)

plt.show()

```

```python
from scipy.stats import spearmanr

# sample data
x = ans['score'][0:15]
y = spat['score'][0:15]

# calculate Spearman's correlation coefficient and p-value
corr, pval = spearmanr(x, y)

# print the result
print("Spearman's correlation coefficient:", corr)
print("p-value:", pval)
```
